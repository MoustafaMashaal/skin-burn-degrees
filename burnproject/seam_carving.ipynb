```python
from matplotlib import pyplot as plt
from skimage.transform import resize as rs
import cv2


def resize(image):
    carved = rs(image, (100, 100))
    return carved
```

```python
img1 = cv2.imread("burn.jpg", 3)          
rgb_img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB) 
plt.figure()
plt.title('experimenting with seam carving')
plt.imshow(rgb_img1)
```

```python
plt.imshow(resize(rgb_img1))
```
